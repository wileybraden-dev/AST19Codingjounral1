```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
plt.rc("font", size=18)
#this cell is just importing numpy and matplotlib as their respective shorthands for latter use. It also reformats the text in the plots for a better viewing. 
#started 2/27/2026, finished __/__/2026
```

```python
x = np.linspace(0,2*np.pi, 100)
#this cell is jsut defining x as a serries of 100 numbers going from 0 to 2pi
```

```python
def exponent(x,a = 4 - np.pi, b = 1.5):
    y = a*np.exp(-b*x)
    return(y)
# In this cell I  made the function "exponent" which defined y as the equation given in the prompt, and defined a and b for the equation y
```

```python
plt.plot(x, exponent(x,a = 4 - np.pi, b = 1.5))
plt.xlabel("x")
plt.ylabel("y")
plt.savefig(fname = "expgraph.png")
# This cell is just ploting x and y and giving the two axises labells, not too much. 
```

```python
sigma = np.linspace(2,.2,100)
scatter = np.random.normal(scale = sigma, size = 100)
print(scatter)
#This does almost exactly what the promt said, make a list of 100 points linearly going from 2 to .2 using np.linspace, and using np.random.normal to make the scatter. 
```

```python
plt.plot(x, exponent(x,a = 4 - np.pi, b = 1.5), color="blue", zorder = 1)
plt.errorbar(x, exponent(x,a = 4 - np.pi, b = 1.5), yerr = np.absolute(scatter), color ="red", zorder=0)
plt.xlabel("x")
plt.ylabel("y + scatter")
plt.savefig(fname = "expgraph.png")
#This cell is almost identical to the prior graph, having the same underlying graph which looks kind of squished, but using the errorbar style of graph and making the y error the scatter
#To make the core graph stand out I tried to make it a different color, but I couldn't figure that out so I just coppied the core graph and put it on top and made it blue.
```

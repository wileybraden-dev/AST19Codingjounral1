```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
plt.rcParams["font.family"] = "Times New Roman"
plt.rc("font", size=18)
#this cell is just importing numpy and matplotlib as their respective shorthands for latter use. It also reformats the text in the plots for a better viewing. 
#started 2/27/2026, finished 5/11/2026
```

```python
x = np.linspace(0,2*np.pi, 100)
#this cell is jsut defining x as a serries of 100 numbers going from 0 to 2pi
```

```python
def exponent(x,a = 10 - np.pi, b = 1.5):
    y = a*np.exp(-b*x)
    return(y)
# In this cell I  made the function "exponent" which defined y as the equation given in the prompt, and defined a and b for the equation y
```

```python
plt.plot(x, exponent(x,a = 10 - np.pi, b = 1.5))
plt.xlabel("x")
plt.ylabel("y")
plt.savefig(fname = "expgraph.png")
# This cell is just ploting x and y and giving the two axises labells, not too much. 
```

```python
sigma = np.linspace(2,.2,100)
scatter = np.random.normal(scale = sigma, size = 100)
print(scatter)
#This does almost exactly what the promt said, make a list of 100 points linearly going from 2 to .2 using np.linspace, and using np.random.normal to make the scatter. 
```

```python
plt.errorbar(x, exponent(x,a = 10 - np.pi, b = 1.5)+scatter, color ="red", zorder=0)
plt.xlabel("x")
plt.ylabel("y + scatter")
plt.savefig(fname = "expgraph_error.png")
#This cell is almost identical to the prior graph, having the same underlying graph which looks kind of squished, but adding in the scatter
#the phasing on the question is sort of baffling to me, becuase it wants to use the errorbar graph but says to add the scatter to the y value directly, so there isn't an errorbar, or at least that is my reading of it.
```

```python
def model_1(x,a1,b1):
    y_model_1 = a1*np.exp(-b1*x)
    return y_model_1

model_1_test = curve_fit(model_1, x, exponent(x,a = 10 - np.pi, b = 1.5) + scatter)
print(model_1_test)
plt.plot(x,model_1(x,a1 = 5.81388449, b1 = 1.23180356), zorder = 1, label="model 1")
plt.plot(x, exponent(x,a = 10 - np.pi, b = 1.5) + scatter, label="data")
plt.savefig(fname = "expgraph_model1.png")
plt.legend()
#This cell defines the function "model_1" which is the exponential from the prompt
#Then it uses curve_fit to find the best values and plots them out, along with the data, and has a legend and saves the graph
#The curve fit fitted the function pretty well
```

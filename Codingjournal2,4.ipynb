```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
plt.rc("font", size=18)
#this cell is just importing numpy and matplotlib as their respective shorthands for latter use. It also reformats the text in the plots for a better viewing. 
#started 5/11/2026, finished 5/11/2026
```

```python
y = np.random.wald(1,.8,100)
print(y)
#This cell is just defining y as a np.random.wald function, with a mean and sclae that I thought looked interesting, and 100 values.
```

```python
plt.hist(y, bins = 100)
plt.xlabel("x")
plt.ylabel("y")
plt.savefig(fname = "Hist.png")
#this cell plots the y data as a histogram with 100 bins, and also labels the x and y axises and saves the graph.
```
